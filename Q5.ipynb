# **Question 5 : Vectorized Method**

```
root = "/content/drive/My Drive/2018AAPS1242H_NNFL (Assignment 1)/"
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

%matplotlib inline

training_output = pd.read_excel(root+"Data/training_output.xlsx",header=None)
test_output = pd.read_excel(root+"Data/test_output.xlsx",header=None)
training_feature_matrix = pd.read_excel(root+"Data/training_feature_matrix.xlsx",header=None)
test_feature_matrix = pd.read_excel(root+"Data/test_feature_matrix.xlsx",header=None)
```

```
learning_rate = 0.4
num_iterations = 50
regularization_parameter = 1
batch_size = 35
```

```
X = np.array(training_feature_matrix)
X = np.concatenate((np.ones(len(X)).reshape(-1,1), X.reshape(-1,2)), axis=1)
y = np.array(training_output)

X = X/np.linalg.norm(X)
y = y/np.linalg.norm(y)
w = np.zeros((X.shape[1], 1))
```

```
def vectorized_linearRegression(X, y):
    w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
    return w
```

```
def predict(X, w):
    return X @ w
```

```
w = vectorized_linearRegression(X, y)
```

```
test_matrix = np.array(test_feature_matrix)
test_matrix = np.concatenate((np.ones(len(test_matrix)).reshape(-1,1), test_matrix.reshape(-1,2)), axis=1)
```

```
y_pred = predict(test_matrix, w)
y_pred = y_pred/np.linalg.norm(y_pred)
```

```
y_test = np.array(test_output)
y_test = y_test/np.linalg.norm(y_test)
```

```
def error(y_pred, y_output):
    error = 0
    for i in range(y_pred.shape[0]):
        error += (y_pred[i]-y_output[i])**2   
    return str(error.reshape(-1,))
```

```
error_ = error(y_pred, y_test)
```

```
print("MSE error in Prediction is " + error_)
```
